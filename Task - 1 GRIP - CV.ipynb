### Task 1 : Optical Character Recognition ( Computer Vision )

```python
#Importing libraries
import cv2
import matplotlib.pyplot as plt
import easyocr
from pylab import rcParams
import warnings
warnings.filterwarnings("ignore")
from IPython.display import Image
rcParams['figure.figsize'] = 10,20
```

```python
#Downloading English Language Model
Lang_reader = easyocr.Reader(['en'])
```

```python
Image('thought.jpg')
```

### Test in Image - thought.jpg 

```python
output1 = Lang_reader.readtext('thought.jpg')
```

### Print the detected text  

```python
output1
```

```python
cord1 = output1[-2][0]
print(output1[-2][1])
```

```python
x_min1,y_min1 = [min(idx) for idx in zip(*cord1)]
```

```python
x_max1,y_max1 = [max(idx) for idx in zip(*cord1)]
```

```python
image = cv2.imread('thought.jpg')
cv2.rectangle(image,(int(x_min1),int(y_min1)),(int(x_max1),int(y_max1)),(0,255,0),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
